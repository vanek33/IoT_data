## WSU SmartApartment 
#### Описание
    Набор представляет собой данные с датчиков в квартире собранные в период с 2010 по 2012 годы. В этот период в квартире проживало двое жильцов. Выполнялась повседневная деятельность. Случалось что жильцы покидали квартиру в период с 9 до 17 по будним дням. 
    Набор представляет собой данные в едином текстовом файле. Действия участников (жильцов) не привязаны к данным.

#### Описание переменных
Данные с датчиков можно классифицировать по их названию:
   Mxxx: датчик движения
   MAxxx: датчик зоны движения (покрывает большую область)
   Ixxx: датчик предметов для выбранных предметов на кухне
   Dxxx: датчик двери
   AD1-A: датчик горелки
   AD1-B: датчик горячей воды
   AD1-C: датчик холодной воды
   Винтаж: датчики температуры
   Р001: потребление электроэнергии 
   Baxxxx: уровень заряда батареи
   LLxxx: уровень освещенности 
   SSxxx: датчик встряхивания
   LSxxx: датчик освещенности
   SGxxx: датчик гироскопа
   Exxx: переключатель экспериментатора (ручной пуск)
   Fxxx: вентилятор
